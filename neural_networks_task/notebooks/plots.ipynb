```python
import torch.nn.functional as F
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results_enet = pd.read_csv("../results/EfficientNetV2_50_32_1e-05/results.csv")
results_mnet = pd.read_csv("../results/MobileNetV3_50_32_1e-05/results.csv")
```

```python
results_mnet.columns
```

```python
sns.lineplot(data=results_enet, x="epoch", y="train_accuracy", label="EfficientNet Train Accuracy")
sns.lineplot(data=results_mnet, x="epoch", y="train_accuracy", label="MobileNet Train Accuracy")
```

```python
sns.lineplot(data=results_enet, x="epoch", y="eval_accuracy", label="EfficientNet Evaluation Accuracy")
sns.lineplot(data=results_mnet, x="epoch", y="eval_accuracy", label="MobileNet Evaluation Accuracy")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the metrics you want to plot
weighted_metrics = [
    "train_weighted_recall",
    "train_weighted_precision",
    "train_weighted_f1",
    "eval_weighted_recall",
    "eval_weighted_precision",
    "eval_weighted_f1",
]

# Set up the matplotlib figure
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10), sharex=True)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through the metrics and create a line plot for each
for i, metric in enumerate(weighted_metrics):
    sns.lineplot(ax=axes[i], data=results_enet, x="epoch", y=metric, label="EfficientNet")
    sns.lineplot(ax=axes[i], data=results_mnet, x="epoch", y=metric, label="MobileNet")
    axes[i].set_title(metric.replace("_", " ").title())
    axes[i].legend(loc="upper left")

# Adjust layout for better fit and to prevent labels from overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
